```python
import neurokit2 as nk
import matplotlib.pyplot as plt
```

```python
nk.fractal_mandelbrot(show=True)
```

```python
m = nk.fractal_mandelbrot(real_range=(-2, 0.75), imaginary_range=(-1.25, 1.25))
plt.imshow(m.T, cmap="viridis")
plt.axis("off")
plt.show()
```

```python
b = nk.fractal_mandelbrot(size=1500,
                          real_range=(-2, 0.75), imaginary_range=(-1.25, 1.25),
                          buddha=True, iterations=200)
plt.imshow(b.T, cmap="gray")
plt.axis("off")
plt.show()
plt.savefig("buddhabrot_grey.png", dpi=300)
```

```python
plt.imshow(b.T, cmap="twilight_shifted")
plt.axis("off")
plt.show()

plt.savefig("buddhabrot_twiglight.png", dpi=300)
```

```python
plt.imshow(b.T, cmap="bone")
plt.axis("off")
plt.show()
```

```python
plt.imshow(b.T, cmap="binary")
plt.axis("off")
plt.show()
```
